### 总体概述
该应用程序是一个视频播放器，支持在视频播放的时候，对当前播放的视频进行截图操作. 

###  开发框架与技术选型
使用 Gradio 5.0 来快速搭建交互式的Web界面


### 功能描述

应用程序叫视频元数据增强器。主要有两大功能，元数据增强的视频理解 和 交互式图文搜索视频。

1. 应用程序的标题是：视频元数据增强器：让视频内容发挥价值
2. 程序包括两个table页，一个是元数据增强，一个是交互式图文搜索

3. 元数据增强的视频理解界面包括两个部分，

    3.1 元数据增强的视频理解界面左侧

        3.1.1 提供文本输入框包含输入让用户输入要播放的视频链接信息

        3.1.2 链接下方是开始播放按钮，播放按钮下方是视频播放区域，点击按钮后，播放器根据链接播放视频。 

        3.1.3 在播放器下方左侧，有手动分析分组，手动分析包括模型选择框和开发分析按钮
        
        3.1.4 在播放器下方左侧，有自动分析分组，手动分析包括模型选择框，分析时间，和开始分析按钮

    3.2 元数据增强的视频理解界面右侧包含，视频名称，视频ID，视频描述文本框，标签选择列表和视频截图展示区

4. 交互式图文搜索视频界面包括

    1. 上传图片区域

    2. 高美感图片选项

    3. 交互式对话框
    
    4. 搜索结果展示区，展示视频列表

### 增加byteplus sdk 的调用
1.把视频链接改成视频空间，用户输入视频空间名称回车后，自动从从后台或获取一个视频列表，视频列表是一个价值对，包括Vid和Title。获取视频列表参考byteplus_sdk_function.
2. 增加下拉框选项，获取视频列表后在列表中展示获取的视频列表；
3. 点击开始播放后，从获取的视频列表获取vid信息，进行视频播放。
